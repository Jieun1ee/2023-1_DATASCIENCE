```python
df['month'] = df['신고일시'].dt.month
df['day'] = df['신고일시'].dt.weekday
```

```python
monthCount = dict(df['month'].value_counts())
monthCount = dict(sorted(monthCount.items()))
monthCount
```

```python

mon=monthCount.keys()
singo = monthCount.values()
plt.rc('font', family='Malgun Gothic') # For Windows
plt.figure(figsize=(11,6))
plt.title('월 별 신고 건수')
plt.xticks([i+1 for i in range(12)])
plt.plot(mon, singo)
```

```python
seasons = {0:0, 1:0, 2:0, 3:0}
seasons
```

```python
for k, v in monthCount.items():
    if 3<=k<=5 :
        seasons[0] += v
    if 6<=k<=8 :
        seasons[1] += v        
    if 9<=k<=11 :
        seasons[2] += v
    if k==12 or 1<=k<=2 :
        seasons[3] += v
```
